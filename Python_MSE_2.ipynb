**№10** Напишите программу, которая находит сумму всех чисел от 1 до 100 и выводит ее на экран. Получите плюс к моральному удовлетворению за то, что следовали Гауссу

```python
i = 1
answer = 0
while i <= 100:
  answer += i # то же самое, что answer = answer + i
  i += 1 # это тоже самое, что i = i + 1
  print("Сейчас i = ", i)
  print("Сейчас answer = ", answer)
print("Сумма = ", answer)
print("Плюс к моральному удовлетворению")
```

```python
i = 1
answer = 0
while i <= 100:
  answer += i # то же самое, что answer = answer + i
  i += -1 # это тоже самое, что i = i + 1
  print("Сейчас i = ", i)
  print("Сейчас answer = ", answer)
  if i < -1000 or i > 1000:
    print("Вышли из цикла через break")
    break

print("Вышли из цикла")
```

# Цикл while

**№14** Напишите программу, которая запрашивает у пользователя число, а затем выводит на экран таблицу умножения для этого числа от 1 до 10.

```python
n = int(input("Введите число: "))
i = 1
while i <= n:
  print(f"{i} * {n} = {i * n}")
  i += 1
```

**№18** \* Напишите программу, которая запрашивает у пользователя число, а затем выводит на экран сумму его цифр.

```python
n = int(input("Введите число: "))
answer = 0
while n > 0:
  answer += n % 10
  n = n // 10
print(answer)
```
