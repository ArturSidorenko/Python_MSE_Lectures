# Функции

Функция --- кирпичек, из которых строится большая программа, продпрограмма, выполняющая свою небольшую задачу.

Вы встречались с такими функциями, как `print()` и `input()`.

Функции позволяют группировать код, что решает проблему дублирования кода.

```python
print('Hello!')
print('I am fine, thanks')
print('Hello!')
print('I am fine, thanks')
print('Hello!')
print('I am fine, thanks')
```

Свою функцию можно сделать с помощью ключевого слова `def`

```python
def greet():
  print('Hello!')
  print('How are u?')

for j in range(3):
  greet()
```

В функцию можно передавать параметры (аргументы). Типа данных аргумента не определяется, в отличие от других языков программирования. Вы можете написать `name: str` и заявить, что `name` --- строка (string). Но это ДЛЯ ПОЛЬЗОВАТЕЛЯ, не для Python




```python
def say_hello(name: str):
  print(f'Hello, {name}')

say_hello('Bob')
say_hello(True)
```

**Основные термины**

*   Определить функцию --- это создать ее
*   Аргументы функции --- те данные, которые мы передаем в функцию
*   Возвращаемое значение --- то, что дает функция по завершении, задается оператором `return`


```python
def square(a: int):
  return a * a

print(square(5))
print(square(2.5))
```

Здесь мы познакомимся с модулем `random`, он отвечает за генерацию случайных чисел. Этот модуль имеет функцию ` randint`, которая генерирует целые числа.

```python
import random

def toss_coin():
  n = random.randint(1, 2)
  if n == 1:
    return 'орел'
  else:
    return 'решка'
```

**№1** Написать функцию `is_even()`, которая принимает число в качестве аргумента и возвращает True, если число чётное, и False, если число нечётное.

```python
def is_even(a):
  if a % 2 == 0:
    return True
  else:
    return False

print(is_even(5), is_even(14))
```

```python
def is_even(a):
  return not bool(a % 2)

print(is_even(5), is_even(14))
```

**№2** Написать функцию `multiply()`, которая принимает два числа в качестве
аргументов и возвращает их произведение.

```python
def multiply(a, b):
  return a * b

print(multiply(3, 12), multiply(-2, 5))
```

**№3** Написать функцию `calculate_power()`, которая принимает два числа в качестве аргументов и возвращает первое число, возведённое в степень второго числа. Второе число натуральное.

```python
def calculate_power(x, a):
  ans = 1
  for i in range(a):
    ans = ans * x
  return ans

calculate_power(5, 3)
```

**№4** Написать функцию `collatz()`, которая принимает натуральное число n вкачестве аргумента и возвращает n // 2, если n четно, и 3 * n + 1, если n нечетно.

```python
def collatz(n):
  if is_even(n):
    return n // 2
  else:
    return 3 * n + 1

n = 29
for j in range(100):
  n = collatz(n)
  print(n)
```

**№5** Написать функцию `is_prime()`, которая принимает число в качестве аргумента и возвращает True, если число простое, и False, если число составное.

```python
def is_prime(n):
  flag = True
  for k in range(2, n // 2):
    if n % k == 0:
      flag = False
      break
  return flag

is_prime(102)
```
